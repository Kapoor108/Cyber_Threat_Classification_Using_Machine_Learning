```python
import numpy as np
import pandas as pd
import matplotlib as plt
```

```python
import pandas as pd

# Load the dataset
file_path = "cybersecurity_dataset.csv"  # Update with your file path
df = pd.read_csv(file_path)

# Display basic information
print("Dataset Info:")
print(df.info())

# Display the first few rows
print("\nFirst 5 rows of the dataset:")
print(df.head())
```

```python
# Drop unnecessary columns
df_cleaned = df.drop(columns=["id", "timestamp", "src_ip", "dst_ip"])

# Verify the dataset after dropping columns
print("\nDataset after dropping unnecessary columns:")
print(df_cleaned.head())
```

```python
from sklearn.preprocessing import LabelEncoder

# Encode categorical feature (protocol)
encoder = LabelEncoder()
df_cleaned["protocol"] = encoder.fit_transform(df_cleaned["protocol"])

# Verify encoding
print("\nUnique protocol values after encoding:")
print(df_cleaned["protocol"].unique())
```

```python
from sklearn.preprocessing import StandardScaler

# Initialize the scaler
scaler = StandardScaler()

# Normalize numerical columns
df_cleaned[["bytes_sent", "bytes_received", "flow_duration"]] = scaler.fit_transform(
    df_cleaned[["bytes_sent", "bytes_received", "flow_duration"]]
)

# Verify normalization
print("\nDataset after normalization:")
print(df_cleaned.head())
```

```python
# Save the preprocessed dataset
df_cleaned.to_csv("preprocessed_data.csv", index=False)

print("\nPreprocessed data saved as 'preprocessed_data.csv'")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Compute correlation matrix
corr_matrix = df_cleaned.corr()

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Feature Correlation Matrix")
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier

# Define features and target variable
X = df_cleaned.drop(columns=["is_attack"])
y = df_cleaned["is_attack"]

# Train a basic Random Forest model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X, y)

# Extract feature importance
feature_importance = pd.DataFrame({
    "Feature": X.columns,
    "Importance": rf_model.feature_importances_
}).sort_values(by="Importance", ascending=False)

# Plot feature importance
plt.figure(figsize=(8, 5))
sns.barplot(x=feature_importance["Importance"], y=feature_importance["Feature"], palette="viridis")
plt.title("Feature Importance (Random Forest)")
plt.xlabel("Importance Score")
plt.ylabel("Feature")
plt.show()
```

```python
from sklearn.feature_selection import mutual_info_classif

# Compute Mutual Information scores
mi_scores = mutual_info_classif(X, y, discrete_features=False)

# Create a DataFrame for better visualization
mi_df = pd.DataFrame({"Feature": X.columns, "MI Score": mi_scores})
mi_df = mi_df.sort_values(by="MI Score", ascending=False)

# Plot MI scores
plt.figure(figsize=(8, 5))
sns.barplot(x=mi_df["MI Score"], y=mi_df["Feature"], palette="plasma")
plt.title("Mutual Information Scores for Features")
plt.xlabel("MI Score")
plt.ylabel("Feature")
plt.show()
```

```python
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier

# Initialize model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)

# Apply RFE
rfe = RFE(rf_model, n_features_to_select=5)  # Select top 5 features
X_rfe = rfe.fit_transform(X, y)

# Get selected features
selected_features = X.columns[rfe.support_]

print("Selected Features using RFE:", selected_features)
```

```python
from sklearn.model_selection import train_test_split

# Define features (X) and target (y)
X_final = df_cleaned[selected_features]  # Use selected features from Feature Selection
y = df_cleaned["is_attack"]

# Split dataset (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X_final, y, test_size=0.2, random_state=42, stratify=y)

print(f"Training Set: {X_train.shape}, Testing Set: {X_test.shape}")
```

```python
from sklearn.ensemble import RandomForestClassifier

# Initialize model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)

# Train model
rf_model.fit(X_train, y_train)

# Make predictions
y_pred_rf = rf_model.predict(X_test)
```

```python
from sklearn.svm import SVC

# Initialize SVM model
svm_model = SVC(kernel="rbf", random_state=42)

# Train model
svm_model.fit(X_train, y_train)

# Make predictions
y_pred_svm = svm_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report

# Evaluate Random Forest
rf_accuracy = accuracy_score(y_test, y_pred_rf)
rf_report = classification_report(y_test, y_pred_rf)

# Evaluate SVM
svm_accuracy = accuracy_score(y_test, y_pred_svm)
svm_report = classification_report(y_test, y_pred_svm)

print("Random Forest Performance:\n", rf_report)
print("SVM Performance:\n", svm_report)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

# Plot Confusion Matrix for Random Forest
plt.figure(figsize=(6, 5))
sns.heatmap(confusion_matrix(y_test, y_pred_rf), annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix - Random Forest")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()

# Plot Confusion Matrix for SVM
plt.figure(figsize=(6, 5))
sns.heatmap(confusion_matrix(y_test, y_pred_svm), annot=True, fmt="d", cmap="Reds")
plt.title("Confusion Matrix - SVM")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV

# Define parameter grid
param_grid_rf = {
    "n_estimators": [50, 100, 200],
    "max_depth": [10, 20, None],
    "min_samples_split": [2, 5, 10],
}

# Initialize GridSearchCV
grid_rf = GridSearchCV(RandomForestClassifier(random_state=42), param_grid_rf, cv=5, scoring="accuracy", n_jobs=-1)

# Train the model
grid_rf.fit(X_train, y_train)

# Get best parameters
best_rf = grid_rf.best_estimator_
print("Best Random Forest Parameters:", grid_rf.best_params_)

# Evaluate optimized model
y_pred_rf_best = best_rf.predict(X_test)
print("Optimized Random Forest Performance:\n", classification_report(y_test, y_pred_rf_best))
```

```python
# Define parameter grid
param_grid_svm = {
    "C": [0.1, 1, 10],
    "kernel": ["linear", "rbf"],
    "gamma": ["scale", "auto"]
}

# Initialize GridSearchCV
grid_svm = GridSearchCV(SVC(random_state=42), param_grid_svm, cv=5, scoring="accuracy", n_jobs=-1)

# Train the model
grid_svm.fit(X_train, y_train)

# Get best parameters
best_svm = grid_svm.best_estimator_
print("Best SVM Parameters:", grid_svm.best_params_)

# Evaluate optimized model
y_pred_svm_best = best_svm.predict(X_test)
print("Optimized SVM Performance:\n", classification_report(y_test, y_pred_svm_best))
```
